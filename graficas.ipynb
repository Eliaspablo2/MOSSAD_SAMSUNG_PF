```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Estadsticas de Casos Sometidos.csv')
df['Cantidad'] = pd.to_numeric(df['Cantidad'], errors='coerce')
df = df.dropna(subset=['Cantidad'])


# Grfico de barras
# Comparacin de la cantidad de delitos entre provincias
delitos_por_provincia = df.groupby('Provincia')['Cantidad'].sum()
umbral_delitos = 50000 
provincias_con_mas_delitos = delitos_por_provincia[delitos_por_provincia > umbral_delitos]
plt.figure(figsize=(20, 10))  # Ajusta el tamao de la figura segn sea necesario
provincias_con_mas_delitos.sort_values(ascending=False).plot(kind='bar')
plt.title('Provincias con ms de 10,000 Delitos')
plt.xlabel('Provincia')
plt.ylabel('Cantidad de Delitos')
plt.xticks(rotation=45)
plt.show()







# Grfico de lneas:
# Tendencia temporal de la cantidad de delitos
plt.figure(figsize=(10, 6))
df['Fecha'] = df['Mes'] + '-' + df['Ao'].astype(str)
df.groupby('Fecha')['Cantidad'].sum().plot(kind='line', marker='o')
plt.title('Tendencia Temporal de la Cantidad de Delitos')
plt.xlabel('Fecha')
plt.ylabel('Cantidad de Delitos')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()





# GRAFICO PASTEL SANTIAGO
df['Cantidad'] = pd.to_numeric(df['Cantidad'], errors='coerce')
df = df.dropna(subset=['Cantidad'])
provincia_seleccionada = 'Santiago'
df_provincia = df[df['Provincia'] == provincia_seleccionada]
delitos_por_tipo = df_provincia.groupby('Delito')['Cantidad'].sum()
umbral_delitos = 5000
delitos_con_mil_casos_o_mas = delitos_por_tipo[delitos_por_tipo >= umbral_delitos]
plt.figure(figsize=(8, 8))
delitos_con_mil_casos_o_mas.plot(kind='pie', autopct='%1.1f%%')
plt.title(f'Proporcin de Delitos en {provincia_seleccionada} con ms de 5000 casos')
plt.ylabel('')
plt.show()



# GRAFICO PASTEL SANTO DOMINGO
df['Cantidad'] = pd.to_numeric(df['Cantidad'], errors='coerce')
df = df.dropna(subset=['Cantidad'])
provincia_seleccionada = 'Santo domingo'
df_provincia = df[df['Provincia'] == provincia_seleccionada]
delitos_por_tipo = df_provincia.groupby('Delito')['Cantidad'].sum()
umbral_delitos = 2000
delitos_con_mil_casos_o_mas = delitos_por_tipo[delitos_por_tipo >= umbral_delitos]
plt.figure(figsize=(8, 8))
delitos_con_mil_casos_o_mas.plot(kind='pie', autopct='%1.1f%%')
plt.title(f'Proporcin de Delitos en {provincia_seleccionada} con ms de 2000 casos')
plt.ylabel('')
plt.show()
```
